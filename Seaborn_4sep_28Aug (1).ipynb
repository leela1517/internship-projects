```python
sns.relplot(data=data , x='petal_length',y = 'sepal_width',hue='species',kind='line',ci=None)
```

#Categorical plots

------> it is used to analyse one numerical and categorical column

1. count plot

2. box plot

3. bar plot

4. violin plot

5. strip plot

6. swarm plot

7. cat plot

```python
#Count plot

#--> it just gives the count of categories

sns.countplot(data=data,x ='species')
```

```python
#bar plot

#one num, one cat

sns.barplot(data=data,x='species',y='petal_width')
```

```python
#box plot

#box plot is used to identidy outliers

sns.boxplot(data=data,x = 'petal_length')
```

```python
sns.boxplot(data=data,x='species',y='sepal_length')
```

```python
sns.boxplot(data=data, x ='sepal_width')
```

```python
#violin plot

sns.violinplot(data=data,x='sepal_width')
```

```python
sns.violinplot(data=data,x='species',y='sepal_width')
```

```python
#strip plot

sns.stripplot(data=data,x='species',y='sepal_width',hue='species')
```

```python
#swarmplot

sns.swarmplot(data=data,x='species',y='sepal_width',hue='species')
```

```python
#cat plot

#using cat plot we can draw box,viloin,strip,swarm

sns.catplot(data=data, x='species',y='sepal_length',hue='species',kind='box')
```

```python
#distribution plot

#hist
#dist

#these two graphs are used to check distribution of data
```

```python
sns.distplot(x = data['sepal_length'])
```

```python
sns.distplot(x = data['sepal_width'])
```

```python
sns.histplot(x= data['sepal_length'],kde = True) #kde --> kernel density estimation ---> kde is approxizmation of histogram to density curve
```

```python
#pair plot

sns.pairplot(data,hue='species')
```
